# Browser Notebook Version

This notebook uses the existing `detector.py` model logic and displays image results inline in a Jupyter browser session.

Run the cells below to load an image, detect ships, and show bounding boxes without the `tkinter` desktop UI.

```python
from pathlib import Path
from IPython.display import display
from PIL import Image, ImageDraw

from detector import ShipDetector

# Use the notebook working directory as app root, matching the repository layout.
APP_ROOT = Path('.').resolve()
IMAGES_DIR = APP_ROOT / 'data' / 'images'
IMAGE_EXTS = {'.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp'}

if not IMAGES_DIR.exists():
    raise FileNotFoundError(f'Expected image folder not found: {IMAGES_DIR}')

images = sorted([p for p in IMAGES_DIR.iterdir() if p.suffix.lower() in IMAGE_EXTS])
if not images:
    raise RuntimeError(f'No supported image files found in {IMAGES_DIR}')

images[:10]
```

```python
def draw_boxes(image, boxes, color='red', width=4):
    output = image.copy()
    draw = ImageDraw.Draw(output)
    for x1, y1, x2, y2 in boxes:
        draw.rectangle([x1, y1, x2, y2], outline=color, width=width)
    return output
```

```python
detector = ShipDetector()
detector
```

```python
# Change this index to inspect a different image from data/images.
image_index = 0
image_path = images[image_index]

pil_image = Image.open(image_path).convert('RGB')
boxes = detector.detect(image_path)

result_image = draw_boxes(pil_image, boxes)
display(result_image)
print(f'Detected {len(boxes)} ship(s) in {image_path.name}')
```
